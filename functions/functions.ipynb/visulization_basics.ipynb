```python
x = np.arange(-10, 10, 0.1)
print(x.shape)
```

```python
x2 = x ** 2 # square of x
```

```python
plt.plot(x,x2)
```

```python
plt.plot?
```

```python
x3 = x ** 3
```

```python
plt.figure(figsize=(7,10))
plt.plot(x,label ='original',linestyle='--')
plt.plot(x2, label='squared',linestyle=':')
plt.plot(x3, label='cubed')
plt.legend()
plt.xlabel("Data Points")
plt.ylabel("value of x")
plt.title('The visulization of X and its square and cube') # add a title to the figure
plt.show() # show the figure
```

```python
cities = ['lucknow','delhi','mumbai','pune','bangalore']
rainfall = [39,20,62,45,78]
```

```python
plt.bar(cities,rainfall)
```

```python
plt.bar?
```

```python
plt.barh(cities, rainfall)
```

```python
plt.figure(figsize=(10,7))
plt.bar(cities, rainfall, color='#ff8800', edgecolor='black', hatch='//', width=.5)
plt.xlabel("Cities name")
plt.ylabel("maximum rainfall")
plt.title("Cities rainfall graph")
```

```python
plt.figure(facecolor='white')
wedges = [0,.3,0,0,0]
plt.pie(rainfall, labels=cities, autopct='%1.1f%%',radius=2,explode=wedges)
plt.show()
```

```python
prices = [213,34,46,678,45,67,5,78,34,56,23]
qty= [12,45,67,89,34,23,12,45,23,56,78]
quality = [13,45,68,89,3,41,2,3,12,33,45]
len(prices) == len(qty) == len(quality)
```

```python
plt.scatter(prices, qty)
```

```python
plt.scatter(prices, qty,color='green', marker='s',s=100*3,alpha=.5)
plt.scatter(prices, quality,color='red', marker='o',s=100*2,alpha=.5,edgecolors='black')
```

```python
x =np.random.randint(1,10,500)
y =np.random.randint(1,10,500)
plt.hist(x,bins = 10, color='green',range=(1,10),align='left') # use bins to clear portion of chart
plt.hist(y,bins = 10, color='blue',range=(1,10),align='left',alpha=.5) # use bins to clear portion of chart
_= plt.xticks(np.arange(0,11,1))
```

```python
plt.hist2d(x,y,bins=10,cmap='Blues')
plt.colorbar()
plt.show()
```

```python
plt.boxplot([prices,qty,quality])
_ =plt.xticks(np.arange(1,4),['prices','qty','quality'],rotation=45)
```
