```python
years =[ '1980', '1981', '1982', '1983', '1984', '1985', '1986', '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013']
```

```python
canada['total'] =canada.loc[:, years].sum(axis=1) # axis1 is column , loc using change the values
canada.head(3)
```

```python
canada['mean']= canada.loc[:,years].mean(axis=1)
canada.head(3)
```

```python
canada.rename(
    columns={
        'OdName':'Country',
        'AreaName':'Continent',
        'RegName':'Region',
        'DevName':'Status',
    },
    inplace=True #permanent something we can use inplace 
)
```

create  a new column that stores sum of imigration from 1980-1990
create  a new column that stores sum of imigration from 1990-2000
create  a new column that stores sum of imigration from 2000-2010


```python
canada['1980-1989']=canada.loc[:,map(str,range(1980,1990))].sum(axis=1)
canada['1990-1999']=canada.loc[:,map(str,range(1990,1999))].sum(axis=1)
canada['2000-2010']=canada.loc[:,map(str,range(2000,2010))].sum(axis=1)
canada.head()
```

```python
decade_cols = ['1980-1989','1990-1999','2000-2009']
plt.figure(facecolor='white')
canada.loc['Japan', decade_cols].plot(kind='pie',figsize=(10,6), autopct='%1.1f%%')
```

```python
country_list = canada.index.tolist()
```

```python
canada.loc[country_list[:10],decade_cols].T.plot(
    kind='pie',
    figsize=(50,50),
    autopct='%1.1f%%',
    subplots=True,
    layout=(2,5),
)
plt.show()
plt.savefig('first_five_countries.png',bbox_inches='tight')
```

```python
canada.groupby('Continent').sum()[years].T.plot(kind='line',figsize=(10,6))
```
