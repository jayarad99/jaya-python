create a list of size n,with random numbers from 0 to 100

```python
from random import randint

n = int(input("Size of List:")) # size of list
x = []                          # empty list
for i in range(n):
    val = randint(1,100)    # random num between 1 and 100
    x.append(val)  # add value to list
print(x)    
```

generate a list of 25 elements then create smaller list of number from that based on 
. number that is divisible by 3
. num that is dived 5
num that is divide by both 3 ans 5


```python
x = list(range(26))
print(x)
xs = []
for i in x:
    if i % 3 == 0 and i % 5 == 0:
        xs.append(i)
    elif i % 3 == 0:
        xs.append(i)  
    elif i % 5 == 0:
         xs.append(i)  
print(xs)         
```

take names of 10 movies from user & store them in a list,then display the list in a sorted order

```python

movie = []
for uc in range(10):
    uc = input("Enter movie name :")
    movie.append(uc)
movie.sort()
print(movie)
```
